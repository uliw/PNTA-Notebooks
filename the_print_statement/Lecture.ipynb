## Printing with print()



### Avoiding the linefeed



    print("The chemical formula of", end="")
    print("Barite is")



### Adding it explicitly with an escape sequence



    print("The chemical formula of \nBarite is")



### Typical escape sequences



Note, these are not python commands, rather, these escape sequences
are part of the ASCII standard:


| Escape Sequence|Meaning|
|---|---|
| <code>\newline</code>|Ignored|
| <code>\</code>|Backslash (\)|
| <code>\'</code>|Single quote (')|
| <code>\"</code>|Double quote (")|
| <code>\a</code>|ASCII Bell (BEL)|
| <code>\b</code>|ASCII Backspace (BS)|
| <code>\f</code>|ASCII Formfeed (FF)|
| <code>\n</code>|ASCII Linefeed (LF)|
| <code>\r</code>|ASCII Carriage Return (CR)|
| <code>\t</code>|ASCII Horizontal Tab (TAB)|
| <code>\v</code>|ASCII Vertical Tab (VT)|
| <code>\xA1</code>|ASCII character with hexadecimal value A1|
| <code>\135</code>|ASCII character with octal value 135|



#### The meaning of \\ as escape sequence



    print("Include a linefeed with \new line")
    print("Include a linefeed with \\new line")

the backslash is used to **escape** the meaning of the backslash!

    print("Include a linefeed with "new" line")

    print("Include a linefeed with \"new\" line")



## f-strings simplify your life



#### old school



    fd = {"Barite" : "BaSO4"}
    print("The chemical formula of Barite is ",end="")
    print(fd["Barite"])



#### f-strings



    fd = {"Barite" : "BaSO4"}
    print(f"The chemical formula of Barite is {fd['Barite']}")



#### Example



    a = 2
    b = 4
    c = a * b
    print(f"Multiplying {a} with {b} yields {c}")



## Format modifiers



f-strings provide a convenient way to control formatting 

    m = 2**64
    print(f"{m:4.4e}")

    print(f"{1/3:.2f}")
    print(f"{1/3 * 1000:.2f}")



### padding



    print(f"{1/3:0.2f}")
    print(f"{1/3:0>8.2f}")
