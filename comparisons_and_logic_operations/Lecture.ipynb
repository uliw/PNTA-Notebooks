## Logic Operators



### And/Or



```python
print(False or True)
print(False and True)
print(False and False)
```

### Logical or (xor)



also **exclusive or**, i.e., only one of the conditions can be true
Consider the regular **or** :



```python
print(f"False or False = {False or False}")
print(f"False or True = {False or True}")
print(f"True or True = {True or True}")
```

**xor**, python syntax `a ^ b`



```python
print(f"False xor False = {False ^ False}")
print(f"False xor True = {False ^ True}")
print(f"True xor True = {True ^ True}")
```

### Invert (also `not`)



```python
print(not True)
print(False and not True)
```

### Binary Addition with boolean operators



    # always start on the right moving to the left
    
      0101 # = 5
    + 0010 # = 2
    ------------
      0111 # = 7 -> xor allows binary addition
    
    
    # --- example with carry over --- #
      0101 # = 5
    + 0101 # = 5
    ------------
      0000 # xor
      0000 # and - but 1 + 1 needs a the carry over bit
      1010 # carry over bits
      1010 = 10 



### Binary Subtraction with boolean operators



    # always start on the right moving to the left
    
      0101 # = 5
    - 0010 # = 2
    ------------
    
    1) Invert the number 2 by flipping bits
       0010 = 2
    -> 0001 = 2 inverted 
    
    2) add one bit to the right 
       0001 
       0001 + 1 consider carry over if necssary
    -> 0010 
    
    3) add both numbers with xor, ignoring the left most bit
       0101 = 5
    +  0010 = 2
    -----------
    -> 0011 = 3
