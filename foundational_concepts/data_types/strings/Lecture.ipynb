#### Functions



Functions require an argument, and return a copy of the argument

    my_list = [5, 1, 3]
    new_list = sorted(my_list)
    #                  ^--- argument to function
    #           ^---------- function name
    #  ^------------------- replace my_list with return value of function
    
    print(my_list)
    print(new_list)



#### Methods



Methods belong to an object instance, and often modify the object data,so that the original data is lost.

    my_list = [5, 1, 3] # create an instance of the list object 
    my_list.sort()
    new_list = my_list
    #                   ^--- method name
    #            ^---------- object instance name
    #  ^-------------------- New name for object instance reference to my_list!
    
    print(my_list)
    print(new_list)



#### Functions and methods can be arguments to other functions and methods



    # In this example, sorted() and print() are two functions
    my_list = [5, 1, 3]
    my_sorted_list = sorted(my_list)
    print(my_sorted_list)
    print(sorted(my_list))



#### Key-Value arguments



Both functions and methods often accept additional arguments in the form of a key-value pair(s)

    my_list = [5, 1, 3]
    my_list.sort(reverse=True)
    print(my_list)

#### Unicode



-   ASCII has 256 entries (1 byte!)
-   contains only the American alphabet, and no special characters!
-   required all sorts of extensions, often country and operating system depend.
-   1991, introduction of Unicode,
-   Widespread Unicode adoption with the advent of the internet around the turn of the century
-   UTF8 = 2 to 4 8 bit words,  used universally, fully compatible with ASCII
-   UTF16 = 2 x 16-bit words, not compatible with with ASCII, often buggy, used only by Microsoft.
-   UTF32 = 4 x bit words fixed length -> huge memory footprint, rarely used.
