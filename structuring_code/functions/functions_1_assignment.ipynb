#### Building your own items() function 6 points



The `items()` method of a dictionary returns a list of tuples with the dictionary key, and its value for each dictionary entry.



```python
my_dict = {
    "Name": "John Doe",
    "Phone #": "645 934 1212",
}

print(my_dict.items())
```

```python
my_dict = {
    "Name": "John Doe",
    "Phone #": "645 934 1212",
}

print(my_items(my_dict))
```
