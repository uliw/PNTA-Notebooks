## Block Statements (Code groups)



### Single branch



code groups are delineated by

-   a grouping statement followed by a colon
-   a group of python statements that are indented to the right



```python
a = 10

if a < 12:
    print("This branch is executed when the 'if statement' is True")
    print("a is smaller than 12\n")
 
else:
    print("This branch is executed when the 'if statement' is False")
    print("a is larger (or equal) than 12\n")

print("This part of the code is outside the block")
print("and is always executed")
```

### Multiple branches



-   You can have, one branch, two way branching, and multiple branches.
-   However, no more than one branch will be executed
-   The first branch that is valid will take precedence
-   When coding, always test all branches, by choosing suitable values.



```python
voltage =2
if voltage < 2:
    print("Your sample is to small")
elif voltage > 2 and voltage < 20:
    print(f"Voltage = {voltage}V")
elif voltage > 20 and voltage < 50:
    print(f"Your sample is to big")
elif voltage > 50:
    print("---------------------------------")
    print(f"\n\nAttention: Instrument malfunction!\n\n")
    print("---------------------------------")
else:
    print(f"You should never see this")
```

### Nested blocks (branches)



code blocks can be nested:



```python
voltage = 0.1
if voltage < 50:
    if voltage <= 20:
        if voltage > 2:
            print(f"Voltage = {voltage}V")
        else:
            print(f"Your sample is to small")
    else:
        print(f"Your sample is to big")
else:
    print("---------------------------------")
    print(f"\n\nAttention: Instrument malfunction!\n\n")
    print("---------------------------------")
```

### Pitfalls



```python
a = 12
if a > 5:
    print("a is bigger than 5")
elif a > 7:
    print("a is also bigger than 7")
else:
    print(f"a={a}")
```

### Loop statements create code groups



```python
ml = "Hello World"

for e in ml:
    print(e)

print("\ndone with the loop")
```

### Understanding loops



#### Loops



```python
a = "Hello glorious world".split(" ")

for e in a:
    print(e)
    
""" this is equivalent to
e = a[0]
print(e)

e = a[1]
print(e)

e = a[2]
print(e)
"""
```

#### Counters



```python
a = "Hello glorious world"
c = 0
for e in a:
    c = c + 1
    print(f"{e}, {c}")

""" This is equivalent to
e = a[0]  #  H
c = c + 1 # = 1
H, 1

e = a[1]  # e
c = c + 1 # = 2
e, 2

e = a[2]  # l
c = c + 1 # = 3
l, 3
"""
```

#### Conditionals



```python
a = "Hello glorious world"
c = 0
for e in a:
    if c == 3:
        print(e)
    c = c + 1
```

#### While loops



```python
c = 0
while c < 3:
    print(c)
    c = c + 1

""" is equivalent to
if c < 3: # c = 0
    print(c)
    c = c + 1 # c = 1

if c < 3: # c = 1
    print(c)
    c = c + 1 # c = 2

if c < 3: # c = 2
    print(c)
    c = c + 1 # c = 3

if c < 3: # c = 3 -> False
# code stops here
"""
```

### Explicit looping (e.g., matlab)



```python
ml = "Hello World"
n = len(ml)

for numbers in range(0,n,1):
    print(numbers)
```

### Implicit Looping the python way



```python
ml = "Hello World"

for e in ml[0::2]:
    print(e)
```

### Loops with a running index



```python
ml = "Hello World"

for x, y in enumerate(ml):
    print(f"The letter at position {x} equals '{y}'")
```

### Combining loops and if statements



```python
# counting specific letters in a string
ltc: str = "o"
ml: str = "Hello World"
c: int = 0  # We will use this for counting

for e in ml:  # loop over each string element
    if e == ltc:  # find matching string elements
        c = c + 1  # increase counter by 1

# print result
print(f"The letter '{ltc}' occurs {c} times in {ml}")
```

### Modifying list elements in a loop



Will this work?



```python
# remove specific numbers from a list
ltc: int = 2
ml: list = [1, 2, 4, 3, 2, 5, 8, 2]

cml: list = ml.copy()
for i, e in enumerate(cml):  # loop over each string element
    if e == ltc:  # find matching string elements
        print(i)
        ml.pop(i)

# print result
print(ml)
```

### How to keep track of things



-   print is your friend:



```python
i: int = 0
a: bool = True
b: list = [1, 2, 4]

while a < 10 :
    print(b[i])
    i = i + 1
```

-   use print statements to follow your code execution:



```python
i: int = 0
a: bool = True
b: list = [1, 2, 4]

while a < 10 :
    print(f"a = {a}, i ={i}")
    print(f"b[i] = {b[i]}\n")
    i = i + 1
```
