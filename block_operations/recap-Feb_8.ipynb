## Recap



### If statements create code groups (blocks)



    a = 12
    
    if a == 12:
        print("a is exactly 12")
    elif a < 12:
        print("a is smaller than 12")
    elif a > 12:
        print("a is larger than 12")
    else:
        print("This should never happen")



### Loop statements create code groups



    ml = "Hello World"
    
    for e in ml:
        print(e)
    
    print("done with the loop")



### Explicit loops



    for e in range(5):
        print(e)



### Loops and counters



    i = 0
    for e in range(5):
        i = i + 2
        print(i)



### Loop with index counter



    s = "Hello World"
    i = 0 
    for e in s[:-1]:
        print(f"i = {i}, e = {e} s[{i}+1] = {s[i+1]}")
        i = i + 1



### enumerate



    s = "ab"
    print(list(enumerate(s)))



### tuple unpacking



    a = (1, 2)
    x, y = a
    print(f"x = {x}, y = {y}")



### Loop with automatic index



    s = "Hello World"
    
    for i, e in enumerate(s[:-1]):
        print(f"i = {i}, e = {e} s[{i}+1] = {s[i+1]}")



### While loops



    a = True
    i = 0
    while a:
        print(i)
        if i > 3:
            break
        else:
            i = i + 1



### Looping over a dictionary



    d = {"A": 12, "B": 0}
    
    for key, value in d.items():
        print(f"key = {key}, value = {value}")
