These arrays are list-like objects that resemble pandas Series. However, unlike `pd.Series`, these arrays do not include metadata like an index. Pay attention to the type hinting used for the NumPy array object. We will cover this topic in detail during the NumPy module, but for now, please adhere to this example.



```python
import numpy as np
import numpy.typing as npt  # import numpy type hints

# declare the type hint for a float array
NDArrayFloat = npt.NDArray[np.float64]

# generate a NumPy array filled with values from
# zero to 2*pi with a step size of 0.1
x_1: NDArrayFloat = np.arange(0, np.pi * 2, 0.1)
y_1: NDArrayFloat = np.sin(x_1)  # calculate the sin(x)
y_2: NDArrayFloat = np.cos(x_1)  # calculate the cos(x)
y_3: NDArrayFloat = np.cos(x_1) ** 2 + 2  # calculate the cos(x)^2 + 2
y_4: NDArrayFloat = y_3 + +y_2 + y_1  # calcualte the sum y1 to y3
```
