### Background



The reconstruction of Earth's thermal history relies heavily on the ability to extract precise environmental signals from the fossil record. Among the most successful tools developed for this purpose is the magnesium-to-calcium (Mg/Ca) ratio in the calcitic tests of foraminifera. Since its emergence as a quantitative paleothermometer in the late 20th century, this proxy has transformed our understanding of ocean temperature variability, global ice volume, and carbon cycle dynamics. The theoretical basis for the Mg/Ca proxy is rooted in the endothermic substitution of $\ce{Mg^2+}$ for $\ce{Ca^2+}$ in the calcium carbonate (CaCO<sub>3</sub>) lattice, a process that is thermodynamically favored at higher temperatures. Here we will use the original dataset Marchitto et al 2007 to evaluate the predictive power of the Mg/Ca thermometer.
